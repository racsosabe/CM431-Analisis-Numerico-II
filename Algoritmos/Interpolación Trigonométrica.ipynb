```python
# Plot polynomials

def plotPoly(polys, name, L, R):
    x = var('x')
    a = plot([],figsize=(8, 8),title=name)
    classic_colors = ['crimson', 'blue', 'green', 'orange', 'purple', 'brown', 'red', 'yellow']
    n = len(polys)
    assert(n > 0)
    assert(len(L) == n and len(R) == n)
    for i in range(n):
        a += plot(polys[i], x, L[i], R[i], color=colors.keys()[i])
    show(a)
```

```python
# Parse polynomial with coefficients

def parsePoly(poly):
    var('x')
    ans = 0
    mono = 1;
    for i in range(len(poly.list())):
        ans += poly.list()[i] * mono
        mono *= x
    return ans
```

```python
# Plot polynomials with interpolation method "interpolation"

def plotPolylines(groups, name, interpolation):
    polynomials = []
    L = []
    R = []
    for points in groups:
        p = parsePoly(interpolation(points))
        polynomials.append(p)
        minimum = points[0][0]
        maximum = points[0][0]
        for i in range(1, len(points)):
            if points[i][0] < minimum: minimum = points[i][0]
            if points[i][0] > maximum: maximum = points[i][0]
        L.append(minimum)
        R.append(maximum)
    plotPoly(polynomials, name, L, R)
```

# Algoritmos para Interpolación Trigonométrica

```python
def Gaussian(a, b):
    n = len(a)
    assert(n > 1)
    m = len(a[0])
    assert(m > 1)
    assert(len(b) == n)
    EPS = 1e-15
    ans = [None for _ in range(m)]
    where = [-1 for _ in range(m)]
    row = 0
    col = 0
    while(row < n and col < m):
        sel = row
        for i in range(row, n):
            if abs(a[i][col]) > abs(a[sel][col]):
                sel = i
        if abs(a[sel][col]) < EPS:
            continue
        for i in range(col, m):
            a[sel][i], a[row][i] = a[row][i], a[sel][i]
        b[sel], b[row] = b[row], b[sel]
        where[col] = row
        for i in range(n):
            if i == row: continue
            c = a[i][col] / a[row][col]
            for j in range(col, m):
                a[i][j] -= a[row][j] * c
            b[i] -= b[row] * c
        row += 1
        col += 1
    for i in range(m):
        if where[i] == -1: continue
        ans[i] = b[where[i]] / a[where[i]][i]
    for i in range(n):
        s = 0
        for j in range(m):
            s += a[i][j] * ans[j]
        if abs(s - b[i]) > EPS:
            for k in range(m):
                ans[k] = None
            return ans
    return ans
```

```python
def InterpolacionTrigonometrica(points):
    n = len(points)
    assert(n % 2 == 1)
    l = (n - 1) // 2
    a = [[0 for j in range(n)] for i in range(n)]
    b = [points[i][1] for i in range(n)]
    for i in range(n):
        a[i][0] = 1
        for j in range(1, l + 1):
            a[i][j] = cos(points[i][0] * j)
        for j in range(1, l + 1):
            a[i][j + l] = sin(points[i][0] * j)
    A = Gaussian(a, b)
    P = A[0]
    var('x')
    for i in range(1, l + 1):
        P += cos(i * x) * A[i]
    for i in range(1, l + 1):
        P += sin(i * x) * A[i + l]
    return P
    
```

# Ejemplo

```python
points = [(4.0, 20.0), (5.0, 30.0), (7.0, 60.0)]

P = InterpolacionTrigonometrica(points)

print(P)
plotPoly([P], "Interpolacion Trigonometrica", [4], [7])
```
