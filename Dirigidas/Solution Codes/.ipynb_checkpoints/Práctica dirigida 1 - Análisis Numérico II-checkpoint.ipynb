```python
R = PolynomialRing(QQ, 'x')
def plotPoly(polys, name, l, r):
    x = var('x')
    a = plot([],figsize=(5,5),title=name)
    classic_colors = ['crimson', 'blue', 'green', 'orange', 'purple', 'brown', 'red']
    n = len(polys)
    for i in range(n):
        print("Adding polynomial :", polys[i], "with color", classic_colors[i])
        a += plot(polys[i], x, l[i], r[i], color=classic_colors[i])
    show(a)
```

```python
def Lagrange(points):
    #x = R.0
    n = len(points)
    assert(n > 1)
    var('x')
    ans = 0
    for i in range(n):
        act = points[i][1]
        for j in range(n):
            if i == j: continue
            act *= (R.gen() - points[j][0]) / (points[i][0] - points[j][0])
        ans += act
    return ans
```

```python
def Newton(points):
    #x = R.0
    n = len(points)
    assert(n > 1)
    var('x')
    memo = matrix(QQ, n, n, 0)
    for i in range(n):
        memo[i, 0] = points[i][1]
    for l in range(1, n):
        for i in range(0, n - l):
            memo[i, l] = (memo[i+1][l-1] - memo[i][l-1]) / (points[i+l][0] - points[i][0])
    ans = 0
    prefix = 1
    for i in range(n):
        ans += prefix * memo[0][i]
        if i + 1 < n: prefix *= (R.gen() - points[i][0])    
    return ans
```

```python
def Hermite(points): # FIX THIS
    #x = R.0
    n = len(points)
    sumk = 0
    for x in points:
        sumk += len(x) - 1
    assert(sumk > 1)
    var('x')
    memo = matrix(QQ, sumk, sumk, 0)
    vis = matrix(ZZ, sumk, sumk, 0)
    pos = 0
    X = []
    for x in points:
        k = len(x) - 1
        for p in range(k):
            X.append(x[0])
        r = pos + k
        for i in range(pos, r):
            for l in range(r - i):
                memo[i, l] = x[l+1] / factorial(l)
                vis[i, l] = 1
        pos += k
    for l in range(1, sumk):
        for i in range(0, sumk - l):
            if vis[i, l] == 1: continue
            vis[i, l] = 1
            memo[i, l] = (memo[i+1][l-1] - memo[i][l-1]) / (X[i+l] - X[i])
    ans = 0
    prefix = 1
    for i in range(sumk):
        ans += prefix * memo[0][i]
        if i + 1 < sumk: prefix *= (R.gen() - X[i])
    return ans
```

```python
def parsePoly(poly):
    var('x')
    ans = 0
    mono = 1;
    for i in range(len(poly.list())):
        ans += poly.list()[i] * mono
        mono *= R.gen()
    return ans
```
