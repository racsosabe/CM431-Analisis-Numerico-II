```python
R = PolynomialRing(QQ, 'x')
```

```python
def plotPoly(polys, name, L, R): # Fix This
    x = var('x')
    a = plot([],figsize=(5,5),title=name)
    classic_colors = ['crimson', 'blue', 'green', 'orange', 'purple', 'brown', 'red']
    n = len(polys)
    assert(n > 0)
    assert(len(L) == n and len(R) == n)
    for i in range(n):
        a += plot(polys[i], x, L[i], R[i], color=classic_colors[i])
    show(a)
```

```python
def Lagrange(points):
    #x = R.0
    n = len(points)
    assert(n > 1)
    ans = 0
    for i in range(n):
        act = points[i][1]
        for j in range(n):
            if i == j: continue
            act *= (x - points[j][0]) / (points[i][0] - points[j][0])
        ans += act
    return ans
```

```python
def Newton(points):
    #x = R.0
    n = len(points)
    assert(n > 1)
    memo = matrix(QQ, n, n, 0)
    for i in range(n):
        memo[i, 0] = points[i][1]
    for l in range(1, n):
        for i in range(0, n - l):
            memo[i, l] = (memo[i+1][l-1] - memo[i][l-1]) / (points[i+l][0] - points[i][0])
    ans = 0
    prefix = 1
    for i in range(n):
        ans += prefix * memo[0][i]
        if i + 1 < n: prefix *= (x - points[i][0])    
    return ans
```

```python
def Hermite(points):
    #x = R.0
    n = len(points)
    sumk = 0
    for p in points:
        sumk += len(p) - 1
    assert(sumk > 1)
    memo = matrix(QQ, sumk, sumk, 0)
    vis = matrix(ZZ, sumk, sumk, 0)
    pos = 0
    X = []
    for p in points:
        k = len(p) - 1
        r = pos + k
        for i in range(k):
            X.append(p[0])
        for i in range(pos, r):
            for l in range(r - i):
                memo[i, l] = p[l+1] / factorial(l)
                vis[i, l] = 1
        pos += k
    for l in range(1, sumk):
        for i in range(0, sumk - l):
            if vis[i, l] == 1: continue
            vis[i, l] = 1
            memo[i, l] = (memo[i+1][l-1] - memo[i][l-1]) / (X[i+l] - X[i])
    ans = 0
    prefix = 1
    for i in range(sumk):
        ans += memo[0][i] * prefix
        if i + 1 < sumk: prefix *= (x - X[i])
    return ans
```

```python
def parsePoly(poly):
    var('x')
    ans = 0
    mono = 1;
    for i in range(len(poly.list())):
        ans += poly.list()[i] * mono
        mono *= x
    return ans
```

# Pregunta 6

Dada la siguiente tabla:

|   x  |    0   |    1   | 3     |
|:----:|:------:|:------:|-------|
| f(x) | -1.201 | 0.8204 | 2.253 |

a) Construye el polinomio interpolante $P_{n}(x)$ que surge de la fórmula de Newton.

b) Estudiar el error por redondeo durante las operaciones que se comete al evaluar $P_{n}(2)$.

# Solucion

```python
points = [(0, -1.201), (1, 0.8204), (3, 2.253)]

N = parsePoly(Newton(points))

print(N)
print(N(2))
plotPoly([N], "Polinomio interpolante de Newton", [-0.1], [3.1])
```

```python
def obtenerErrorSuma(x, y, dx, dy):
    return dx + dy

def obtenerErrorMultiplicacion(x, y, dx, dy):
    return x * dy + y * dx

def obtenerErrorDivision(x, y, dx, dy):
    return dx / y + x * dy / y ** 2

def multiplyPolynomial(a, b, da, db):
    s = len(a) + len(b)
    ans = [0 for _ in range(s)]
    dans = [0 for _ in range(s)]
    for i in range(len(a)):
        for j in range(len(b)):
            dans[i + j] += obtenerErrorMultiplicacion(a[i], b[j], da[i], db[j])
            ans[i + j] += a[i] * b[j]
    while len(ans) > 1 and ans[-1] == 0 and dans[-1] == 0:
        ans.pop()
        dans.pop()
    return (ans, dans)

def addPolynomial(a, b, da, db):
    s = max(len(a), len(b))
    ans = [0 for _ in range(s)]
    dans = [0 for _ in range(s)]
    for i in range(s):
        if i < len(a):
            ans[i] += a[i]
            dans[i] += da[i]
        if i < len(b):
            ans[i] += b[i]
            dans[i] += db[i]
    return (ans, dans)

def computeError(p, dp, x, dx):
    n = len(p)
    i = n - 1
    ans = 0
    dans = 0
    while i >= 0:
        dans = obtenerErrorMultiplicacion(ans, x, dans, dx)
        ans *= x
        dans = obtenerErrorSuma(ans, p[i], dans, dp[i])
        ans += p[i]
        i -= 1
    return dans

n = 3
points = [(0, -1.201), (1, 0.8204), (3, 2.253)]

errorNewton = matrix(QQ, n, n, 0)
matrizNewton = matrix(QQ, n, n, 0)

for i in range(n):
    errorNewton[i, 0] = 0.000005
    matrizNewton[i, 0] = points[i][1]

for l in range(1, n):
        for i in range(0, n - l):
            errorNewton[i, l] = obtenerErrorDivision(matrizNewton[i+1][l-1] - matrizNewton[i][l-1], points[i+l][0]- points[i][0], errorNewton[i+1][l-1] + errorNewton[i][l-1],0)
            matrizNewton[i, l] = (matrizNewton[i+1][l-1] - matrizNewton[i][l-1]) / (points[i+l][0] - points[i][0])

print("Errores en la matriz de diferencias divididas")
print(errorNewton)

ans = [0]
dans = [0]
prefix = [1]
dprefix = [0]
for i in range(n):
    cur, dcur = multiplyPolynomial(prefix, [matrizNewton[0][i]], dprefix, [errorNewton[0][i]])
    ans, dans = addPolynomial(ans, cur, dans, dcur)
    if i + 1 < n:
        prefix, dprefix = multiplyPolynomial(prefix, [-points[i][0], 1], dprefix, [0, 0])

print("Error de precision por coeficiente del polinomio de Newton")
print(dans)
print("Error al calcular P(2)")
print(computeError(ans, dans, 2, 0))
```
