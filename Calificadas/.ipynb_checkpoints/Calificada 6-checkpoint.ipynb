# Problema 1

Sea la ecuación de ondas

$$ u_{tt} - 16u_{xx} = 0, 0 < x < 1, t > 0 $$

sujeta a las condiciones:

$$ u(0, t) = u(1, t) = 0, t > 0 $$

$$ u(x, 0) = \sin{(\pi x)}, u_{t}(x, 0), 0 \leq x \leq 1 $$

Utilice el método de las diferencias finitas para aproximar considerando $h = 0.2$, $k = 0.05$

Notemos que:

$$ \alpha = \frac{k}{h} = \frac{0.05}{0.2} = \frac{1}{4} \rightarrow \alpha^{2} = \frac{1}{16} $$
$$ c^{2} = 16 $$

Así que tendremos:

$$ u_{i, j + 1} = 2(1 - \alpha^{2}c^{2})u_{i, j} + \alpha^{2}c^{2}(u_{i - 1, j} + u_{i + 1, j}) - u_{i, j - 1} $$

$$ u_{i, j + 1} = 2(1 - 1)u_{i, j} + 1(u_{i - 1, j} + u_{i + 1, j}) - u_{i, j - 1} $$

$$ u_{i, j + 1} = u_{i - 1, j} + u_{i + 1, j} - u_{i, j - 1} $$

Debido a la forma de las diferencias finitas, el resultado obtenido es estable y converge a la solución de la ecuación diferencial parcial original.

Ahora, como $u_{t}(x, 0) = 0$, tenemos que:

$$ \frac{u_{i, 0 + 1} - u_{i, 0 - 1}}{2k} = 0 $$

Por lo tanto:

$$ u_{i, 1} = u_{i, -1} $$

Considerando $j = 0$ en la ecuación original:

$$ u_{i, 0 + 1} = u_{i - 1, 0} + u_{i + 1, 0} - u_{i, 0 - 1} $$

$$ u_{i, 1} = u_{i - 1, 0} + u_{i + 1, 0} - u_{i, -1} $$

Pero como $u_{i, 1} = u_{i, -1}$:

$$ u_{i, 1} = u_{i - 1, 0} + u_{i + 1, 0} - u_{i, 1} $$

$$ u_{i, 1} = \frac{u_{i - 1, 0} + u_{i + 1, 0}}{2}, 0 < i < \frac{1}{h} $$

```python
# 0 -> 1 / h = 1 / 0.2 = 5
# 0 -> infinity

h = 0.2
k = 0.05

n = 5
m = 10

u = [[0 for j in range(m + 1)] for i in range(n + 1)]

for i in range(1, n):
    u[i][0] = sin(acos(-1) * i * h)

for i in range(1, n):
    u[i][1] = (u[i - 1][0] + u[i + 1][0]) / 2.0

for j in range(2, m + 1):
    for i in range(1, n):
        u[i][j] = u[i - 1][j] + u[i + 1][j] - u[i][j - 1]
    
for i in range(n + 1):
    print(u[i][:])
```

# Problema 2

Sea la ecuación del calor

$$ u_{t} - u_{xx} = 0, 0 < x < 1, t > 0 $$

con condiciones de contorno

$$ u(0, t) = u(1, t) = 0, t > 0 $$

condición inicial:

$$ u(x, 0) = \sin{(\pi x)}, 0 \leq x \leq 1 $$

Donde $h = 0.2$, $k = 0.05$

Notemos que:

$$ c^{2} = 1 $$
$$ \alpha = \frac{kc^{2}}{h^{2}} = \frac{0.05}{0.2^{2}} = 1.25 $$


Así que tendremos:

$$ u_{i, j + 1} = \alpha u_{i - 1, j} + (1 - 2\alpha)u_{i, j} + \alpha u_{i + 1, j} $$

$$ u_{i, j + 1} = 1.25 u_{i - 1, j} - 1.5 u_{i, j} + 1.25 u_{i + 1, j} $$

```python
# 0 -> 1 / h = 1 / 0.2 = 5
# 0 -> infinity

h = 0.2
k = 0.05

n = 5
m = 10

u = [[0 for j in range(m + 1)] for i in range(n + 1)]

for i in range(1, n):
    u[i][0] = sin(acos(-1) * i * h)

for j in range(1, m + 1):
    for i in range(1, n):
        u[i][j] = 1.25 * u[i - 1][j - 1] - 1.5 * u[i][j - 1] + 1.25 *  u[i + 1][j - 1]
    
for i in range(n + 1):
    print(u[i][:])
```

# Problema 4

Clasifique y reduzca a la forma canónica la siguiente ecuación:

$$ u_{xx} + u_{xy} - 2u_{yy} - 2u_{x} + 5u_{y} - 3u = 0 $$

Notemos que:

$$ a = 1, 2b = 1, c = -2, d = -2, e = 5, f = -3 $$

$$ b^{2} - ac = \left(\frac{1}{2}\right)^{2} - 1(-2) = \frac{9}{4} > 0 $$

Por lo tanto, esta ecuación es hiperbólica.

Ahora, debemos plantear un cambio de variables $\xi_{x}$, $\xi_{y}$, $\eta_{x}$, $\eta_{y}$ tales que cumplan:

$$ a\xi_{x} + (b + \sqrt{b^{2} - ac})\xi_{y} = 0 $$
$$ a\eta_{x} + (b - \sqrt{b^{2} - ac})\eta_{y} = 0 $$

Entonces, tendremos:

$$ \xi_{x} + \left(\frac{1}{2} + \sqrt{\frac{9}{4}}\right)\xi_{y} = 0 $$
$$ \eta_{x} + \left(\frac{1}{2} - \sqrt{\frac{9}{4}}\right)\eta_{y} = 0 $$

$$ \xi_{x} + \left(\frac{1}{2} + \frac{3}{2}\right)\xi_{y} = 0 $$
$$ \eta_{x} + \left(\frac{1}{2} - \frac{3}{2}\right)\eta_{y} = 0 $$

Así que nos quedamos con:

$$ \xi_{x} + 2\xi_{y} = 0 $$
$$ \eta_{x} - \eta_{y} = 0 $$
