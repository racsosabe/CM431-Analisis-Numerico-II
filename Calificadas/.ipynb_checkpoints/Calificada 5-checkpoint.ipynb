```python
# Plot polynomials

def plotPoly(polys, name, L, R, col = None):
    x = var('x')
    a = plot([],figsize=(8, 8),title=name)
    classic_colors = ['crimson', 'blue', 'green', 'orange', 'purple', 'brown', 'red', 'yellow']
    n = len(polys)
    assert(n > 0)
    assert(len(L) == n and len(R) == n)
    for i in range(n):
        a += plot(polys[i], x, L[i], R[i], color=col)
    show(a)
```

```python
def Lagrange(points):
    #x = R.0
    var('x')
    n = len(points)
    assert(n > 1)
    ans = 0
    for i in range(n):
        act = points[i][1]
        for j in range(n):
            if i == j: continue
            act *= (x - points[j][0]) / (points[i][0] - points[j][0])
        ans += act
    return ans
```

```python
def plotGraph(points, name, color = None):
    n = len(points)
    assert(n > 1)
    pointsdx = points
    polys = []
    L = []
    R = []
    for i in range(n-1):
        P = Lagrange([pointsdx[i], pointsdx[i+1]])
        polys.append(P)
        L.append(pointsdx[i][0])
        R.append(pointsdx[i+1][0])
    plotPoly(polys, name, L, R, color)
```

# Problema 1

Tenemos:

$$ m\frac{d^{2} x}{d t^{2}} + c\frac{dx}{dy} + kx = mA\omega^{2}\sin{\omega t} $$

Si hacemos el cambio de variable:

$$ x_{1} = x $$

$$ x_{2} = x' $$

Obtenemos:

$$ x_{1}' = x_{2} $$

$$ x_{2}' = A\omega^{2}\sin{\omega t} - \frac{k}{m}x_{1} - \frac{c}{m}x_{2} $$

Y esto lo podemos plantear de manera matricial:

$$ \begin{pmatrix} x_{1}' \\ x_{2}' \end{pmatrix} = \begin{pmatrix} 0 &1 \\ -\frac{k}{m} &-\frac{c}{m} \end{pmatrix} \cdot \begin{pmatrix} x_{1} \\ x_{2} \end{pmatrix} + \begin{pmatrix} 0 \\ A\omega^{2}\sin{\omega t} \end{pmatrix} $$

Por dato del problema:

$$ k = 400 $$
$$ m = 120 $$
$$ w_{n} = \sqrt{\frac{k}{m}} = \sqrt{\frac{400}{120}} = \sqrt{\frac{10}{3}} \approx 1.825742 $$
$$ c = w_{n}\cdot m \rightarrow \frac{c}{m} \approx 1.825742 $$

$$ \begin{pmatrix} x_{1}' \\ x_{2}' \end{pmatrix} = \begin{pmatrix} 0 &1 \\ -\frac{400}{120} &-1.825742 \end{pmatrix} \cdot \begin{pmatrix} x_{1} \\ x_{2} \end{pmatrix} + \begin{pmatrix} 0 \\ A\omega^{2}\sin{\omega t} \end{pmatrix} $$

$$ \begin{pmatrix} x_{1}' \\ x_{2}' \end{pmatrix} = \begin{pmatrix} 0 &1 \\ -3.\hat{3} &-1.825742 \end{pmatrix} \cdot \begin{pmatrix} x_{1} \\ x_{2} \end{pmatrix} + \begin{pmatrix} 0 \\ A\omega^{2}\sin{\omega t} \end{pmatrix} $$

La suma de los valores **truncados** de la matriz de $2 \times 2$ es $0 + 1 - 3 - 1 = -3$

```python
# Hallar los puntos en los primeros 15 s con h = 0.1s
# Runge-Kutta de orden 4

h = 0.1
t = 0
x1 = 0
x2 = 0
it = 150
points = [(t, x1, x2)]

R.<x,y,z> = RR[]

f = -3.333333 * x - 1.825742 * y + 3.384656 * sin(5.817780 * z)

for i in range(1, it + 1):
    m1 = h * x2
    k1 = h * f(x = x1, y = x2, z = t + h)
    m2 = h * (x2 + k1 / 2)
    k2 = h * f(x = x1 + m1 / 2, y = x2 + k1 / 2, z = t + h)
    m3 = h * (x2 + k2 / 2)
    k3 = h * f(x = x1 + m2 / 2, y = x2 + k2 / 2, z = t + h)
    m4 = h * (x2 + k3)
    k4 = h * f(x = x1 + m3, y = x2 + k3, z = t + h)
    x1 += (m1 + 2 * m2 + 2 * m3 + m4) / 6
    x2 += (k1 + 2 * k2 + 2 * k3 + k4) / 6
    t += h
    points.append((t, x1, x2))
    print(t, x1, x2)
```

```python
XvT = [(x[0], x[1]) for x in points]
VvT = [(x[0], x[2]) for x in points]

plotGraph(XvT, "Posicion vs Tiempo", "red")
plotGraph(VvT, "Velocidad vs Tiempo", "green")
```

# Problema 2

Tenemos:

$$ L \frac{di(t)}{dt} + Ri(t) + \frac{q(t)}{C} = E(t), t > 0 $$

Además se cumple que:

$$ q(t) = \int\limits_{0}^{t}i(t)dt + q(0) \rightarrow q'(t) = i(t) $$

Si reemplazamos $i(t) = q'(t)$:

$$ L \frac{d^{2}q(t)}{dt^{2}} + R\frac{dq(t)}{dt} + \frac{q(t)}{C} = E(t), t > 0 $$

Si hacemos el cambio de variable:

$$ q_{1} = q(t) $$
$$ q_{2} = q'(t) $$

Tendremos el sistema:

$$ q_{1}' = q_{2} $$

$$ q_{2}' = \frac{1}{L}E(t) - \frac{R}{L}q_{2} - \frac{1}{LC}q_{1} $$

Que se traduce, de forma matricial, a:

$$ \begin{pmatrix} q_{1}' \\ q_{2}' \end{pmatrix} = \begin{pmatrix} 0 &1 \\ -\frac{1}{LC} &-\frac{R}{L} \end{pmatrix} \cdot \begin{pmatrix} q_{1} \\ q_{2} \end{pmatrix} + \begin{pmatrix} 0 \\ \frac{1}{L}E(t)\end{pmatrix} $$

Ya que $E = 1$ y $L = 200mH = 0.2H$, tendremos:

$$ \begin{pmatrix} q_{1}' \\ q_{2}' \end{pmatrix} = \begin{pmatrix} 0 &1 \\ -500000 &-500 \end{pmatrix} \cdot \begin{pmatrix} q_{1} \\ q_{2} \end{pmatrix} + \begin{pmatrix} 0 \\ 5\end{pmatrix} $$

Por lo tanto, del vector independiente solo tenemos $0$ y $5$.

```python
# Runge-Kutta de orden 4

h = 0.00025
t = 0
x1 = 0
x2 = 0
it = 100
points = [(t, x1, x2)]

R.<x,y,z> = RR[]

f = -500000 * x - 500 * y + 5

for i in range(1, it + 1):
    m1 = h * x2
    k1 = h * f(x = x1, y = x2, z = t + h)
    m2 = h * (x2 + k1 / 2)
    k2 = h * f(x = x1 + m1 / 2, y = x2 + k1 / 2, z = t + h)
    m3 = h * (x2 + k2 / 2)
    k3 = h * f(x = x1 + m2 / 2, y = x2 + k2 / 2, z = t + h)
    m4 = h * (x2 + k3)
    k4 = h * f(x = x1 + m3, y = x2 + k3, z = t + h)
    x1 += (m1 + 2 * m2 + 2 * m3 + m4) / 6
    x2 += (k1 + 2 * k2 + 2 * k3 + k4) / 6
    t += h
    points.append((t, x1, x2))
    print(t, x1, x2)
```

```python
QvT = [(x[0], x[1]) for x in points]
IvT = [(x[0], x[2]) for x in points]

plotGraph(QvT, "Carga vs Tiempo", "red")
plotGraph(IvT, "Corriente vs Tiempo", "green")
```

# Problema 3

Tenemos:

$$ y'' - \left(1 - \frac{x}{5}\right)y = x $$

Según el método de disparo, podemos reordenar la ecuación anterior a:

$$ y'' = \left(1 - \frac{x}{5}\right)y + x $$

Y tomaremos:

$$ p(x) = 0 $$

$$ q(x) = 1 - \frac{x}{5} $$

$$ r(x) = x $$

Para poder expresarla como:

$$ y'' = p(x)y' + q(x)y + r(x) $$

Además tenemos las condiciones $y(1) = 2$ y $y(3) = -1$.

Consideraremos dos problemas de valor inicial:

$$ y_{1}'' = p(x)y_{1}' + q(x)y_{1} + r(x) $$
$$ y_{1}(1) = 2, y_{1}'(1) = 0 $$

$$ y_{2}'' = p(x)y_{2}' + q(x)y_{2} + r(x) $$
$$ y_{2}(1) = 0, y_{2}'(1) = 1 $$

Y nuestra solución final $y$ se construirá de la siguiente manera:

$$ y(x) = y_{1}(x) + \frac{-1 - y_{1}(3)}{y_{2}(3)}y_{2}(x) $$

Deberemos hallar $y_{1}(3)$ y $y_{2}(3)$ para determinar la pendiente solicitada.

Aplicaremos el método de Runge-Kutta sobre cada problema de manera independiente pero con el mismo $h$.

$$ y_{11} = y_{1} $$

$$ y_{12} = y_{1}' $$

Entonces

$$ y_{11}' = y_{12} $$
$$ y_{12}' = p(x)y_{12} + q(x)y_{11} + r(x) $$

De manera matricial:

$$ \begin{pmatrix} y_{11}' \\ y_{12}' \end{pmatrix} = \begin{pmatrix} 0 &1 \\ q(x) &p(x) \end{pmatrix} \cdot \begin{pmatrix} y_{11} \\ y_{12} \end{pmatrix} + \begin{pmatrix} 0 \\ r(x)\end{pmatrix} $$

Reemplazando:

$$ \begin{pmatrix} y_{11}' \\ y_{12}' \end{pmatrix} = \begin{pmatrix} 0 &1 \\ 1 - \frac{x}{5} &0 \end{pmatrix} \cdot \begin{pmatrix} y_{11} \\ y_{12} \end{pmatrix} + \begin{pmatrix} 0 \\ x\end{pmatrix} $$

Se da igual con la función $y_{2}$.

```python
# Hallando Y_1(3)
# Runge-Kutta de orden 4
# Y_11(1) = 2, Y_12(1) = 0

digits = 4

h = 0.01
t = 1
x1 = 2
x2 = 0
it = 200
points1 = [(t, x1, x2)]

R.<x,y,z> = RR[]

f = (1 - z / 5) * x + z

for i in range(1, it + 1):
    m1 = round(h * x2, digits)
    k1 = round(h * f(x = x1, y = x2, z = round(t + h, digits)), digits)
    m2 = round(h * (x2 + k1 / 2), digits)
    k2 = round(h * f(x = round(x1 + m1 / 2, digits), y = round(x2 + k1 / 2, digits), z = round(t + h, digits)), digits)
    m3 = round(h * (x2 + k2 / 2), digits)
    k3 = round(h * f(x = round(x1 + m2 / 2, digits), y = round(x2 + k2 / 2, digits), z = round(t + h, digits)), digits)
    m4 = round(h * (x2 + k3), digits)
    k4 = round(h * f(x = round(x1 + m3, digits), y = round(x2 + k3, digits), z = round(t + h, digits)), digits)
    x1 = round(x1 + (m1 + 2 * m2 + 2 * m3 + m4) / 6, digits)
    x2 = round(x2 + (k1 + 2 * k2 + 2 * k3 + k4) / 6, digits)
    t = round(t + h, 4)
    points1.append((t, x1, x2))
    print(t, x1, x2)
```

```python
# Hallando Y_2(3)
# Runge-Kutta de orden 4
# Y_21(1) = 0, Y_22'(1) = 1

h = 0.01
t = 1
x1 = 0
x2 = 1
it = 200
points2 = [(t, x1, x2)]

R.<x,y,z> = RR[]

f = (1 - z / 5) * x + z

for i in range(1, it + 1):
    m1 = round(h * x2, digits)
    k1 = round(h * f(x = x1, y = x2, z = round(t + h, digits)), digits)
    m2 = round(h * (x2 + k1 / 2), digits)
    k2 = round(h * f(x = round(x1 + m1 / 2, digits), y = round(x2 + k1 / 2, digits), z = round(t + h, digits)), digits)
    m3 = round(h * (x2 + k2 / 2), digits)
    k3 = round(h * f(x = round(x1 + m2 / 2, digits), y = round(x2 + k2 / 2, digits), z = round(t + h, digits)), digits)
    m4 = round(h * (x2 + k3), digits)
    k4 = round(h * f(x = round(x1 + m3, digits), y = round(x2 + k3, digits), z = round(t + h, digits)), digits)
    x1 = round(x1 + (m1 + 2 * m2 + 2 * m3 + m4) / 6, digits)
    x2 = round(x2 + (k1 + 2 * k2 + 2 * k3 + k4) / 6, digits)
    t = round(t + h, 4)
    points2.append((t, x1, x2))
    print(t, x1, x2)
```

```python
pendiente = round(round(-1 - points1[-1][1], 5) / points2[-1][1], 5)
print(pendiente)
pointsY = []
for i in range(len(points1)):
    this_x = points1[i][0]
    this_y = points1[i][1] + pendiente * points2[i][1]
    pointsY.append((this_x, this_y))
plotGraph(pointsY, "Metodo de disparo", "blue")
```
