# Problema 1

```python
def Lagrange(points, x0, ndigits = 13):
    #x = R.0
    n = len(points)
    assert(n > 1)
    ans = 0
    for i in range(n):
        act = round(points[i][1], ndigits)
        for j in range(n):
            if i == j: continue
            act = round(round(act * round(x0 - points[j][0], ndigits), ndigits) / round(points[i][0] - points[j][0], ndigits), ndigits)
        ans = round(ans + act, ndigits)
    return ans
```

```python
def Newton(points, x0, ndigits = 13):
    #x = R.0
    n = len(points)
    assert(n > 1)
    memo = matrix(QQ, n, n, 0)
    for i in range(n):
        memo[i, 0] = round(points[i][1], ndigits)
    for l in range(1, n):
        for i in range(0, n - l):
            memo[i, l] = round(round(memo[i+1][l-1] - memo[i][l-1], ndigits) / round(points[i+l][0] - points[i][0], ndigits), ndigits)
    ans = 0
    prefix = 1
    for i in range(n):
        ans = round(ans + round(prefix * memo[0][i], ndigits), ndigits)
        if i + 1 < n: prefix = round(prefix * round(x0 - points[i][0], ndigits), ndigits)
    return ans
```

```python
points = [(3.83, 30.0), (4.17, 35.5), (4.97, 50.5), (6.06, 75.0), (6.71, 92.0), (7.17, 105.0), (7.51, 115.0), (7.98, 130.0)]

x0 = 4.4
cnt = 0

for i in range(3, len(points) + 1):
    cur_points = points[:i]
    L = Lagrange(cur_points, x0, 13)
    N = Newton(cur_points, x0, 13)
    print("n = %d" % (i - 1))
    print(L.n(digits = 13))
    print(N.n(digits = 13))
    print("")
print("Coincidencias = 5")
```

```python
def LagrangeP(points):
    #x = R.0
    n = len(points)
    assert(n > 1)
    ans = 0
    var('x')
    for i in range(n):
        act = points[i][1]
        for j in range(n):
            if i == j: continue
            act *= (x - points[j][0]) / (points[i][0] - points[j][0])
        ans += act
    return ans

def NewtonP(points):
    #x = R.0
    n = len(points)
    assert(n > 1)
    memo = matrix(QQ, n, n, 0)
    for i in range(n):
        memo[i, 0] = points[i][1]
    for l in range(1, n):
        for i in range(0, n - l):
            memo[i, l] = (memo[i+1][l-1] - memo[i][l-1]) / (points[i+l][0] - points[i][0])
    ans = 0
    prefix = 1
    var('x')
    for i in range(n):
        ans += prefix * memo[0][i]
        if i + 1 < n: prefix *= (x - points[i][0])   
    return ans
```

```python
points = [(3.83, 30.0), (4.17, 35.5), (4.97, 50.5), (6.06, 75.0), (6.71, 92.0), (7.17, 105.0), (7.51, 115.0), (7.98, 130.0)]

L = LagrangeP(points)

N = NewtonP(points)

plotPoly([L], "Lagrange", [3.80], [8])
plotPoly([N], "Newton", [3.80], [8])
```

# Problema 2

```python
from math import *
PI = acos(-1)

def getChebyshevPoints(n, a, b):
    global PI
    new_x = []
    for i in range(n + 1):
        den = 2 * (n + 1)
        angulo = PI * (2 * (n - i) + 1) / den
        new_x.append((b + a) / 2 + ((b - a) / 2) * cos(angulo))
    return new_x
```

```python
points = [(3.83, 30.0), (4.17, 35.5), (4.97, 50.5), (6.06, 75.0), (6.71, 92.0), (7.17, 105.0), (7.51, 115.0), (7.98, 130.0)]

new_points = getChebyshevPoints(4, 0, 60)

for i in range(len(new_points)):
    new_points[i] = round(new_points[i], 2)

print("Puntos de Chebyshev (redondeados a centesimas)")
print(new_points)
    
res = 0
for x in new_points:
    res += round(x, 0)
print(res)
```

# Problema 3

Se puede observar que bastan 3 tablas para reproducir el dibujo.

```python
def Gaussian(a, b):
    n = len(a)
    assert(n > 1)
    m = len(a[0])
    assert(m > 1)
    assert(len(b) == n)
    EPS = 1e-15
    ans = [None for _ in range(m)]
    where = [-1 for _ in range(m)]
    row = 0
    col = 0
    while(row < n and col < m):
        sel = row
        for i in range(row, n):
            if abs(a[i][col]) > abs(a[sel][col]):
                sel = i
        if abs(a[sel][col]) < EPS:
            continue
        for i in range(col, m):
            a[sel][i], a[row][i] = a[row][i], a[sel][i]
        b[sel], b[row] = b[row], b[sel]
        where[col] = row
        for i in range(n):
            if i == row: continue
            c = a[i][col] / a[row][col]
            for j in range(col, m):
                a[i][j] -= a[row][j] * c
            b[i] -= b[row] * c
        row += 1
        col += 1
    for i in range(m):
        if where[i] == -1: continue
        ans[i] = b[where[i]] / a[where[i]][i]
    for i in range(n):
        s = 0
        for j in range(m):
            s += a[i][j] * ans[j]
        if abs(s - b[i]) > EPS:
            for k in range(m):
                ans[k] = None
            return ans
    return ans
```

```python
def plotSpline(polys, name):
    polynomials = []
    L = []
    R = []
    for p in polys:
        polynomials.append(p[0])
        L.append(p[1])
        R.append(p[2])
    plotPoly(polynomials, name, L, R)
```

```python
def CubicSplineNatural(points):
    n = int(len(points)) - 1
    assert(n > 0)
    h = [ points[i+1][0] - points[i][0] for i in range(n)]
    beta = [ (points[i+1][1] - points[i][1]) / h[i] for i in range(n) ]
    v = [ (None if i == 0 else 6 * (beta[i] - beta[i - 1])) for i in range(n)]
    a = [ [0 for _ in range(n - 1)] for _ in range(n - 1) ]
    b = [ None for _ in range(n - 1)]
    a[0][0] = 2 * (h[0] + h[1])
    a[0][1] = h[1]
    for i in range(1, n - 2):
        a[i][i] = 2 * (h[i] + h[i+1])
        a[i][i-1] = h[i]
        a[i][i+1] = h[i+1]
    a[n-2][n-3] = h[n-2]
    a[n-2][n-2] = 2 * (h[n-2] + h[n-1])
    for i in range(n - 1):
        b[i] = v[i+1]
    lamb = Gaussian(a, b) # Solve Linear System Ax = b
    lambdas = [None for _ in range(n + 1)]
    lambdas[0] = 0
    lambdas[n] = 0
    for i in range(1, n):
        lambdas[i] = lamb[i-1]
    A = [points[i][1] for i in range(n)]
    B = [ (beta[i] - h[i] * (2 * lambdas[i] + lambdas[i+1]) / 6) for i in range(n)]
    C = [lambdas[i] / 2 for i in range(n)]
    D = [(lambdas[i+1] - lambdas[i]) / 6 / h[i] for i in range(n)]
    polynomials = []
    var('x')
    for i in range(n):
        cur = A[i] + B[i] * (x - points[i][0]) + C[i] * (x - points[i][0]) ** 2 + D[i] * (x - points[i][0]) ** 3
        polynomials.append([cur, points[i][0], points[i+1][0]])
    return polynomials # Returns polynomials with left and right limits for X
```

```python
points1 = [(0.0 * 3, 8.0), (1.0 * 3, 7.0), (2.0 * 3, 6.5), (3.0 * 3, 8.0)]

P1 = CubicSplineNatural(points1)

plotSpline(P1, "Tabla 1")
```

```python
points2 = [(0.5 * 3, 1.5), (1.0 * 3, 1.0), (2.5 * 3, 8.0), (3.0 * 3, 8.0)]

P2 = CubicSplineNatural(points2)

plotSpline(P2, "Tabla 2")
```

```python
points3 = [(0.5 * 3, 1.5), (1.0 * 3, 2.5), (2.0 * 3, 1.6), (4.0 * 3, 1.5)]

P3 = CubicSplineNatural(points3)

plotSpline(P3, "Tabla 3")
```

```python
P4 = []
P4 = P1
for x in P2:
    P4.append(x)
for x in P3:
    P4.append(x)
plotSpline(P4, "Todas las tablas unidas")
# No es tan precisa, pero de algo se vive :v 
```

```python
# Plot polynomials

def plotPoly(polys, name, L, R):
    x = var('x')
    a = plot([],figsize=(8, 8),title=name)
    classic_colors = ['crimson', 'blue', 'green', 'orange', 'purple', 'brown', 'red', 'yellow']
    n = len(polys)
    assert(n > 0)
    assert(len(L) == n and len(R) == n)
    for i in range(n):
        a += plot(polys[i], x, L[i], R[i], color=colors.keys()[i])
    show(a)
```

```python
def multiply(A, B):
    n = len(A)
    m = len(A[0])
    k = len(B[0])
    assert(m == len(B))
    C = [[0 for i in range(k)] for i in range(n)]
    for i in range(n):
        for j in range(k):
            for p in range(m):
                C[i][j] = round(C[i][j] + round(A[i][p] * B[p][j], 6), 6)
    return C
```

```python
def Gaussian(a, b):
    n = len(a)
    assert(n > 1)
    m = len(a[0])
    assert(m > 1)
    assert(len(b) == n)
    EPS = 1e-6
    ans = [None for _ in range(m)]
    where = [-1 for _ in range(m)]
    row = 0
    col = 0
    while(row < n and col < m):
        sel = row
        for i in range(row, n):
            if abs(a[i][col]) > abs(a[sel][col]):
                sel = i
        if abs(a[sel][col]) < EPS:
            continue
        for i in range(col, m):
            a[sel][i], a[row][i] = a[row][i], a[sel][i]
        b[sel], b[row] = b[row], b[sel]
        where[col] = row
        for i in range(n):
            if i == row: continue
            c = a[i][col] / a[row][col]
            for j in range(col, m):
                a[i][j] = a[i][j] - a[row][j] * c
            b[i] = b[i] - b[row] * c
        row += 1
        col += 1
    for i in range(m):
        if where[i] == -1: continue
        ans[i] = b[where[i]] / a[where[i]][i]
    for i in range(n):
        s = 0
        for j in range(m):
            s += a[i][j] * ans[j]
        if abs(s - b[i]) > EPS:
            for k in range(m):
                ans[k] = None
            return ans
    return ans
```

```python
def AjustePolinomial(points):
    n = len(points)
    Y = [[round(log(points[i][1]), 6)] for i in range(n)]
    M = [[1, round(log(points[i][0]), 6)] for i in range(n)]
    MT = [[1 for i in range(n)], [round(log(points[i][0]), 6) for i in range(n)]]
    a = multiply(MT, M)
    c = multiply(MT, Y)
    print(a)
    print(c)
    b = []
    for i in range(2):
        b.append(c[i][0])
    print(b)
    A = Gaussian(a, b)
    print(A)
    var('x')
    print("a - b: %.10f" % (exp(A[0]) - A[1]))
    return exp(A[0]) * x ^ A[1]
```

# Problema 5

# Cambio de variable de $[0, 30]$ a $[-1, 1]$

$$ mx + c $$

$$ m(0) + c = -1 $$
$$ m(30) + c = 1 $$

$$ c = -1 $$
$$ 30m + c = 1 $$

$$ 30m - 1 = 1 $$

$$ m = \frac{2}{30} $$

$$ c = -1 $$

$$ m + c = \frac{2}{30} - 1 = -\frac{28}{30} = -\frac{14}{15} $$

```python
a = 0
b = 30
print(b - a)
X = [sqrt((3 - 2 * sqrt(6 / 5)) / 7), -sqrt((3 - 2 * sqrt(6 / 5)) / 7), sqrt((3 + 2 * sqrt(6 / 5)) / 7), - sqrt((3 + 2 * sqrt(6 / 5)) / 7)]
W = [(18 + sqrt(30)) / 36, (18 + sqrt(30)) / 36, (18 - sqrt(30)) / 36, (18 - sqrt(30)) / 36]
```

```python
points = [(0, 0), (3, 61.40), (6, 73.13), (9, 70.56), (12, 63.43), (15, 55.18), (18, 47.10), (21, 39.83), (24, 33.42), (27, 27.89), (30, 23.20)]

n = 4
var('t')

I = 0

for i in range(n):
    v = X[i] * (b - a) / 2 + (a + b) / 2
    p = 0
    while p < len(points) and points[p][0] <= v:
        p += 1
    if p - 2 >= 0:
        cur_points = [points[p-2], points[p-1], points[p]]
    else:
        cur_points = [points[p-1], points[p], points[p+1]]
    N = NewtonP(cur_points) # Newton para 3 puntos alrededor, tomando preferiblemente hacia la izquierda
    I += (b - a) / 2 * W[i] * N(v)
    
print(I)
```
